```python

# Generate ISO8601 Date/DateTime String
current_time = datetime.datetime.now()

def get_date_str(date):
    '''This take datetime object and convert into iso String req by ICICI Breeze'''
    current_date_str = date.strftime("%d/%m/%Y")
    iso_date_string = datetime.datetime.now().strptime(current_date_str,"%d/%m/%Y").isoformat()[:10] + 'T05:30:00.000Z'
    return iso_date_string
    
def get_date_time_str(date):
    '''This take datetime object and convert into iso String req by ICICI Breeze'''
    current_date_time_str = date.strftime("%d/%m/%Y %H:%M:%S")
    iso_date_time_string = datetime.datetime.strptime(current_date_time_str,"%d/%m/%Y %H:%M:%S").isoformat()[:19] + '.000Z'
    return iso_date_time_string 
    
```

```python
def signal():
    data['shortEMA'] = data.close.ewm(span=9, adjust=False).mean().round(2)
    data['longEMA'] = data.close.ewm(span=21, adjust=False).mean().round(2)

    print(f"ShortEMA of -2 {data['shortEMA'].iloc[-2]}")
    print(data.iloc[-1])

# Determine the crossover point
    if data['shortEMA'].iloc[-1] > data['longEMA'].iloc[-1] and data['shortEMA'].iloc[-2] > data['longEMA'].iloc[-2]:
        print(f'************Singal Recieved 1*************')
        return 1

    elif data['shortEMA'].iloc[-1] < data['longEMA'].iloc[-1]: #and shortEMA[-3] > longEMA[-3]:
        return 0
    
        
```
