```python
import breeze_client as bc
import plotly.graph_objects as go
import pandas as pd 
```

```python


filename = 'data/NIFTY/NIFTY_2024-05-30_22800_call_1second_2024-05-23.csv'
data = bc.read_csv(filename)
data
```

```python

df = data

# Create candlestick chart
fig = go.Figure(data=[go.Candlestick(x=df.index,
                                     open=df['Open'],
                                     high=df['High'],
                                     low=df['Low'],
                                     close=df['Close'])])

# Add titles and labels
fig.update_layout(title='Stock Price Data',
                  xaxis_title='Date',
                  yaxis_title='Price',
                  xaxis_rangeslider_visible=True)

# Save as HTML file
fig.write_html("candlestick_chart.html")

# Show plot in the browser (optional)
fig.show()
```
