```python


dfs = []
for days in range(0,period+1):
    latest_date = dt.date.today()-dt.timedelta(days=days)
    result = bz.download(symbol,interval,'1d',latest_date=latest_date)
    dfs.append(result)
    clear_output()
    print(f'{round(days/period*100,2)} Completed')


```

```python
df = pd.concat(dfs,ignore_index=True)
df
```

```python
df.sort_values('datetime').reset_index(drop=True,inplace=True)
```

```python
filename = f'data/{symbol}/{symbol}_{interval}_{period}{period_unit}.csv'

try:

    df.to_csv(filename,index=False)

except OSError as e:
    print(e)
    os.mkdir(f'data/{symbol}')
    df.to_csv(filename,index=False)

    
```
