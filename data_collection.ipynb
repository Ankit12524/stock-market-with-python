### To get the data for Options or Future

```python
symbol = 'NESIND'
interval = '1minute'
period = 30
period_unit = 'd'
strike_price=2600
right='call'
expiry_date='30-5-2024'
product_type = 'futures'
```

```python
data = bz.download(symbol,interval,period=str(period)+period_unit,exchange_code='NFO',product_type=product_type,expiry_date=expiry_date,right=right,strike_price=strike_price)
data
```

```python


dfs = []
for days in range(0,period+1):
    latest_date = dt.date.today()-dt.timedelta(days=days)
    result = bz.download(symbol,interval,'1d',latest_date=latest_date,exchange_code='NFO',product_type=product_type,expiry_date=expiry_date,right=right,strike_price=strike_price)
    dfs.append(result)
    clear_output()
    print(f'{round(days/period*100,2)} Completed')


```

```python
df = pd.concat(dfs,ignore_index=True)
df
```

```python
df.sort_values('datetime').reset_index(drop=True,inplace=True)
```

```python
if product_type == 'options':
    filename = f'data/{symbol}/{symbol}_{product_type}_{strike_price}_{expiry_date}_{interval}_{period}{period_unit}.csv'

elif product_type == 'futures':
    filename = f'data/{symbol}/{symbol}_{product_type}_{expiry_date}_{interval}_{period}{period_unit}.csv'
else:

    raise(ValueError(f'product type is neither "options" or "futures". If trying to get stock data use First Fucntion in 2nd Cell this is for futrues or options data'))
try:

    df.to_csv(filename,index=False)

except OSError as e:
    print(e)
    os.mkdir(f'data/{symbol}')
    df.to_csv(filename,index=False)

    
```

```python
pd.read_csv(filename)
```
