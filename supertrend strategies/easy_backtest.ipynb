```python
strike = 24500
right = 'call'
expiry = '2024-07-04'
date = '2024-07-03'


filename = f'F:/Python Projects/stock market/stock-market-with-python/data/NIFTY/{date}/{right}/NIFTY_{expiry}_{strike}_call_1second_{date}.csv.csv'
```

```python
data = bz.read_csv(filename=filename)
data = data[['Close','Open','High','Open','Volume','open_interest']]
data
```

```python
df = data['Close'].resample('30S').ohlc()
df['Volume'] = data['Volume'].resample('1T').sum()
df
```

```python
multiplier = 4.0
length = 12
supertrend = ta.supertrend(df['high'],df['low'],df['close'],length = length,multiplier = multiplier)
df  = df.merge(supertrend,on = df.index,how = 'left')
```

```python
# prompt: make a new column signal in df. when last row of 'supertd_7_3.0 is -1 and current row is 1 then signal column value should be 1. if last row of supertd is 1 and current row is -1 then signal should 1 else 0 or na
super = f'SUPERTd_{length}_{multiplier}'

df['signal'] = np.where((df[super].shift(1) == -1) & (df[super] == 1), 1, 0)
df['signal'] = np.where((df[super].shift(1) == 1) & (df[super] == -1), -1, df['signal'])
df['action'] = np.where(df['signal'] == 1, -df['close'], np.nan)
df['action'] = np.where(df['signal'] == -1, df['close'], df['action'])
df
```

```python
df2 = df[df['action'].notnull()]
df2['entry'] = df2['action'].shift(1)
df2['exit'] = df2['action']
df2['profit'] = df2['exit'] + df2['entry']
```

```python
df2
```

```python
win = len(df2[df2['profit'] > 0])
loss = len(df2[df2['profit'] < 0])
print(f'win: {win}, loss: {loss}')
profit = round(df2.profit.sum(),2)
profit_pct = profit/df2['high'].max()
print(f'Profit: {profit}, Profit% = {profit_pct*100}')
```
